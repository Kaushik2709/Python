```python
import pandas as pd

df = pd.read_excel('data.xlsx')
```

```python
df.columns      # This will display the column names of the DataFrame df.
```

```python
df.iloc[0:5, 0:5] # iloc is used to select specific rows and columns based on their integer position. In this case, it selects the first 5 rows and the first 5 columns of the DataFrame.
```

```python
df[(df["City"] == "Pune") & (df["Gross Amount"] > 30000)] # This will filter the DataFrame to include rows where both conditions are met: the "City" is "Pune" and the "Gross Amount" is greater than 30000.
```

```python
df[(df["City"] == "Pune") | (df["Gross Amount"] > 30000)] # This will filter the DataFrame to include rows where either the "City" is "Pune" or the "Gross Amount" is greater than 30000.
```

```python
df[df["Quantity"]>18] # This will filter the DataFrame to include only rows where the "Quantity" column has a value greater than 18.
```

```python
df[df["Customer Name"].str.startswith("A")] # str.startswith() is used to check if the string in the "Customer Name" column starts with the letter "A". It returns a boolean Series where True indicates that the condition is met.
```

```python
df[~df["Customer Name"].str.startswith("A")]  # ~ is used to reverse the condition
```

```python
df.loc[df["Customer Name"] == "Pooja Mehta","Customer Name"] = "Pooja Sharma"  # loc is used to select specific columns based on a condition
```

```python
df[~df["Customer Name"].str.startswith("A")] # ~ it is used to reverse the condition
```

```python
df.dropna()  # used to remove missing values/empty cells from the DataFrame
```

```python
df.groupby("City")["Gross Amount"].sum()
## So here df.groupby("City") will group the data based on the unique values in the "City" column, and then ["Gross Amount"].sum() will calculate the sum of the "Gross Amount" for each city.
```

```python
df.sort_values("Gross Amount") # This will sort the DataFrame in descending order based on the values in the "Gross Amount" column. The row with the highest "Gross Amount" will appear first.
```
